```python
import cv2
import requests
import matplotlib.pyplot as plt

url = "http://localhost:5000/models/super_resolution"
image_source = "images/space.jpg"
image = cv2.imread(image_source)
res = requests.post(url, json={'image': image.tolist()})
print(f"Origin Resolution: {len(image[0])} X {len(image)}")
print(f"Super Resolution: {len(res.json()['super'][0])} X {len(res.json()['super'])}")
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(res.json()['super'])
```

```python
full_bicubic_image, full_superresolution_image = res.json()['origin'], res.json()['super']
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(30, 15))
ax[0].imshow(full_bicubic_image)
ax[1].imshow(full_superresolution_image)
ax[0].set_title("Origin")
ax[1].set_title("Superresolution")
plt.show()
```
